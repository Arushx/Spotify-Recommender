## Load Data

Loading the `Popular_Spotify_Songs.csv` file.

```r
# Using read_csv from readr (part of tidyverse)
# We use show_col_types = FALSE to suppress the column specification output, 
# but you can remove it to see how columns are parsed.
tryCatch({
  df <- read_csv('Popular_Spotify_Songs.csv', show_col_types = FALSE)
  print("Data loaded successfully!")
}, error = function(e) {
  print(paste("Error:", e$message))
})
```

## Data Overview

```r
# Display the first few rows
head(df)

# Display data structure
glimpse(df)

# Basic summary
summary(df)
```
